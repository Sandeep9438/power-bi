● Power BI Desktop

> Power BI is a business intellihence tool that helps users visualize and analyze data from various sources. It allows us to create interactive reports and dashboards, providing insights into our data.

*   **Power BI Desktop**:

> It's a application where users can connect to various data sources, transform data, create data models, and design interactive reports and dashboards. It's the primary authoring tool for Power BI.

*   **Power BI Service**:

>  this is a cloud-based service where users can publish, share, and collaborate on reports and automate and apply the security and dashboards creared in Power BI Desktop. It provides features for data refresh, security, and access management.

*   **Power BI Mobile**:

> It allow users to view and interact with their Power BI reports and dashboards on the go. It provides a touch-optimized experience for consuming insights.

*   **Power BI Gateway**:

> This is a software component that acts as a bridge, providing a secure data transfer between Power BI Service and on-premises data sources . It allows Power BI Service to access and refresh data that is not directly accessible from the cloud.

**Question 2: Compare the following Power BI visuals:
● Pie Chart vs Donut Chart
● Bar Chart vs Column Chart
When would you prefer one over the other? Give one example for each pair ?**

### Pie Chart vs. Donut Chart

> Both Pie Charts and Donut Charts are used to display parts of a whole, showing the proportion of each category within a single dataset.

>   **Pie Chart:**

> A circle divided into sectors, where each sector represents a proportion of the whole. The size of each slice is proportional to the quantity it represents.

>    **Donut Chart:**

> It's similar to a pie chart, but with a blank center. The center hole can be used to display additional information, such as the total value without obscuring the categorical data.

>    **Prefer Pie Chart** when we need to emphasizze that the categories are parts of a whole and the exact percentages are not the primary focus, but rather the relative size comparison. It's often considered more traditional.

>> **Example:** Showing the distribution of sales across different product categories where the exact total sales figure isn't critical to display in the center.

*   **Prefer Donut Chart** when we want to use the central space for a key performance indicator (KPI) or a total value that provides immediate context to the proportions.

>> **Example:** Displaying the market share of different companies in an industry, with the total market revenue shown in the center of the donut chart. With the help of donut chart it's easier to show.

### Bar Chart and Column Chart

> **Bar Chart:** Displays categories along the veertical axis and values along the horizontal axis.

> **Column Chart:** Displays categories along the horizontal axis and values along the vertical axis.

>  **Prefer Bar Chart** (horizontal bars):

>   when we have many categories, or the category names are long, as horizontal labels are easier to read without rotation.
    
>>   **Example:** Comparing the sales performance of the top 10 salespeople, where the names are long and fit batter on the vertical axis.

>  **Prefer Column Chart** (vertical columns) :
    
>> **Example:** Displaying monthly sales revenue over a year, where the months are clearly displayed along the horizontal axis.

**Question 3 :
Explain the significance of:
● Star schema vs Snowflake schema
● Primary key vs Foreign key in relationships (Power BI)
Why is cardinality important?**


>    **Star Schema:**

>    A simple and widely used dimensional modeling schema.It consists of a central fact table surrounded by multiple dimension tables, resembling a star.It simpler to understand and implement, faster quert performance due to fewer joins, and easier to navigate. But it can lead to some data redundancy in dimension tables.

> **Snowflake Schema:**

>   An extension of the star schema where dimension tables are normalized into multiple related tables.It reduces data redundancy, potentially better data integrity due to higher normalization.But it more complex to design and maintain, and can lead to slower query performance due to a greater number of joins required to retrieve data.

> **Primary Key** is a column (or set of columns) that uniquely identifies each row in a table, ensuring no duplicate or null values. It represents the 'one' side of a relationship.

> **Foreign Key** is a column in one table that refers to the primary key in another table. It establishes a link between these tables, representing the 'many' side of a relationship, and allows for accurate data filtering and aggreggation across our data model.

> **Cardinality** refers to the number of unique values in a column or the nature of the relationship between two tables (one-to-many, one-to-one). It is crucial in Power BI for several reasons:

> **Data Integrity and Accuracy:** Correct cardinality ensures that data relationships are accurately represented, allowing filters and calculations to propagate correctly across tables. This leads to reliable and accurate analytical results.

> **Query Performance:** Power BI's engine optimizes queries based on cardinality. Properly defined cardinality helps the engine work efficiently, leading to faster report loading and interaction.

> **Model Design:** Understanding cardinality is fundamental for designing an efficient and understandable data model. It dictates how tables should be joined and how data will flow through the model, impacting overall report performance and usability.

**Question 4 :
Differentiate between:
● Calculated column vs Measure
Also, define Row context and Filter context with simple examples.**

> In Power BI, both **Calculated Columns** and **Measures** are created using DAX, but they serve distinct purposes.

> **Calculated Column**: is a new column added to out data model, with its values computed row by row during data refresh and stored in memory. It's best for row-level calculations or for creating new attributes that you might want to filter or categorize by, and primarily operates under **row context**.

> **Measure** is a dynamic calculation that aggregates data on the fly at query time, based on the filters applied in your report. Measures are not stored in the data model and are primarily evaluated within the **filter context**, making them ideal for aggregated values like 'Total Sales' or 'Average Profit' that change based on user interactions with visuals.

> **Row Context** is when a DAX expression is evaluated row by row, typically for calculated columns. For example, creating a LineTotal calculated column by multiplying Quantity and UnitPrice will compute the total for each individual row in the table.

> **Filter Context** is the set of filters applied to the data before a measure is calculated, determining which data is visible for aggregation. For instance, a Total Sales measure in a bar chart will sum sales only for the specific region that corresponds to each bar, based on the filters active for that visual element.

**Question 5:
What is the difference between a report and a dashboard in Power BI?**

> **Report:** It's a multi-perspective view of a dataset, with visuals that tell a complete story. Reports can have multiple pages, support detailed analysis with filtering and slicing, and allow drilling down into the data. They are designed for in-depth exploration.

> **Dashboard:** A single-page canvas that displays key metrics and visuals from one or more reports. It's meant to provide a quick, at-a-glance overview of important information. Dashboards are static and offer limited interactivity compared to reports; they are primarily for monitoring and quick insights.
